```python
#Big data generation and here 100 random number are generated between 0 and 5
import numpy
x = numpy.random.uniform(0,5,100)
print(x)
```

```python
#Big data generation and show as a histogram
import matplotlib.pyplot as plt
import numpy
x = numpy.random.uniform(0,5,100)
plt.hist(x,5)
#0-1 is taken as one block and 1-2 as 2nd block
plt.show
```

```python
#Big data generation and show as a histogram
import matplotlib.pyplot as plt
import numpy
x = numpy.random.uniform(0,5,10000)
plt.hist(x,100)
#0-1 is taken as one block and 1-2 as 2nd block
plt.show
```
